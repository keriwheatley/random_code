```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))

filename = 'clip_12am.wav'


# read audio samples
input_data = read(filename)
sampling_rate = input_data[0]

print("Sampling Rate:",sampling_rate)

audio = input_data[1]

min_min = 26
max_min = 27

plt.plot(audio[min_min*60*sampling_rate:max_min*60*sampling_rate:22050])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```
